```python
from pyspark.sql import SparkSession
from pyspark.sql import functions as F
import os

def get_col(options, df_columns):
    """
    Helper to find which duration column exists in the current DataFrame.
    """
    for opt in options:
        if opt in df_columns:
            return F.col(opt)
    # Fallback to a literal null if none found to prevent the script from crashing
    return F.lit(None)

def main():
    # 1. Initialize Local Spark
    # .master("local[*]") tells Spark to use all available cores on your laptop
    spark = SparkSession.builder \
        .appName("Duration_Tracker_Local") \
        .master("local[6]") \
        .getOrCreate()
    
    # Set log level to WARN to stop the 'chatter' in your terminal
    spark.sparkContext.setLogLevel("WARN")
    
    # 2. Path Setup
    # Assumes your script is in the parent folder of 'data/raw'
    input_path = "data/raw/*.csv"
    
    if not os.path.exists("data/raw"):
        print("❌ Error: The folder 'data/raw' was not found in your current directory.")
        return

    # 3. Read the Data
    # 'header=True' handles the first row; 'inferSchema=False' keeps everything as String 
    # for the Regex to work safely (Polars-style logic)
    #df_raw = spark.read.option("header", "true").csv(input_path)
    df_raw = spark.read.option("header", "true").option("mergeSchema", "true").csv(input_path)

    dur_opts = ["duration", "Duration", "Total duration", "Duration_Seconds", "duration_seconds"]
    
    # 4. Create a Trace Table
    # We pass df_raw.columns to our helper to find the right column name
    actual_col = get_col(dur_opts, df_raw.columns)

    trace_df = df_raw.select(
        F.input_file_name().alias("source_file"),
        actual_col.alias("raw_string")
    ).withColumn(
        # Extract minutes (e.g., '1h 5m' -> '5')
        "regex_mins", F.regexp_extract(F.col("raw_string"), r"(\d+)\s*m", 1).cast("int")
    ).withColumn(
        # Extract seconds (e.g., '5m 30s' -> '30')
        "regex_secs", F.regexp_extract(F.col("raw_string"), r"(\d+)\s*s", 1).cast("int")
    ).withColumn(
        # If the string is just a number (Old data), cast it directly
        "raw_numeric", F.when(~F.col("raw_string").contains("m"), F.col("raw_string").cast("int"))
    ).withColumn(
        "final_calc", 
        F.when(F.col("raw_numeric").isNotNull(), F.col("raw_numeric"))
        .otherwise((F.coalesce(F.col("regex_mins"), F.lit(0)) * 60) + F.coalesce(F.col("regex_secs"), F.lit(0)))
    )

    # 5. Output
    print("\n" + "="*50)
    print("--- 🕵️ DATA TRACE: Old Data (Numeric Format) ---")
    trace_df.filter(~F.col("raw_string").contains("m")).select("source_file", "raw_string", "final_calc").show(5, truncate=False)

    print("--- 🕵️ DATA TRACE: New Data (String Format: 1h 5m) ---")
    trace_df.filter(F.col("raw_string").contains("m")).select("source_file", "raw_string", "final_calc").show(5, truncate=False)

    total_rows = trace_df.count()
    print(f"✅ TOTAL ROWS PROCESSED: {total_rows}")
    print("="*50 + "\n")

    spark.stop()

if __name__ == "__main__":
    main()
```

```python
from pyspark.sql.functions import col, to_timestamp, hour, year, month, dayofmonth, trim

# 1. Read and Clean
df = spark.read.parquet("data/clean/*.parquet")

# Cast duration and TRIM the start_time to remove hidden spaces
df = df.withColumn("duration_seconds", col("duration_seconds").cast("long")) \
       .withColumn("start_time_raw", trim(col("start_time")))

# 2. Convert to Timestamp
# Using a different name temporarily to verify conversion
df = df.withColumn(
    "start_dt", 
    to_timestamp(col("start_time_raw"), "dd/MM/yyyy HH:mm")
)

# 3. Filter out rows that failed to parse (Prevent downstream errors)
df = df.filter(col("start_dt").isNotNull())

# 4. Derive features with distinct names
df = df.withColumn("start_year", year("start_dt")) \
       .withColumn("start_month", month("start_dt")) \
       .withColumn("start_day", dayofmonth("start_dt")) \
       .withColumn("start_hour", hour("start_dt"))

df.show(5)

# Now your select/orderBy will work perfectly:
df.select("start_year").distinct().orderBy("start_year").show()
```
