```python
import requests
import xml.etree.ElementTree as ET

url = "https://s3-eu-west-1.amazonaws.com/cycling.data.tfl.gov.uk?list-type=2"

response = requests.get(url)
root = ET.fromstring(response.text)

# remove namespace
for elem in root.iter():
    elem.tag = elem.tag.split('}')[-1]

files = []

for item in root.findall(".//Contents"):
    key = item.find("Key").text
    # print(f'item:{item}')
    # print(f'key:{key}')
    if key.endswith(".csv") and "JourneyDataExtract" in key:
        files.append(f"https://cycling.data.tfl.gov.uk/{key}")

print(len(files))
f=files[-1]
print(f)
```

```python
from pyspark.sql.functions import col, to_timestamp, hour, year, month, dayofmonth, trim

# 1. Read and Clean
df = spark.read.parquet("data/clean/*.parquet")

# Cast duration and TRIM the start_time to remove hidden spaces
df = df.withColumn("duration_seconds", col("duration_seconds").cast("long")) \
       .withColumn("start_time_raw", trim(col("start_time")))

# 2. Convert to Timestamp
# Using a different name temporarily to verify conversion
df = df.withColumn(
    "start_dt", 
    to_timestamp(col("start_time_raw"), "dd/MM/yyyy HH:mm")
)

# 3. Filter out rows that failed to parse (Prevent downstream errors)
df = df.filter(col("start_dt").isNotNull())

# 4. Derive features with distinct names
df = df.withColumn("start_year", year("start_dt")) \
       .withColumn("start_month", month("start_dt")) \
       .withColumn("start_day", dayofmonth("start_dt")) \
       .withColumn("start_hour", hour("start_dt"))

df.show(5)

# Now your select/orderBy will work perfectly:
df.select("start_year").distinct().orderBy("start_year").show()
```
