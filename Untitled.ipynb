```python
import requests
import xml.etree.ElementTree as ET

url = "https://s3-eu-west-1.amazonaws.com/cycling.data.tfl.gov.uk?list-type=2"

response = requests.get(url)
root = ET.fromstring(response.text)

# remove namespace
for elem in root.iter():
    elem.tag = elem.tag.split('}')[-1]

files = []

for item in root.findall(".//Contents"):
    key = item.find("Key").text
    # print(f'item:{item}')
    # print(f'key:{key}')
    if key.endswith(".csv") and "JourneyDataExtract" in key:
        files.append(f"https://cycling.data.tfl.gov.uk/{key}")

print(len(files))
f=files[-1]
print(f)
```

```python
import pyarrow.parquet as pq
from pathlib import Path

folder = Path("data/clean")

for file in folder.glob("*.parquet"):
    schema = pq.read_schema(file)
    
    dtype = schema.field("duration_seconds").type
    
    if str(dtype) != "int64":
        print("BAD FILE:", file)
        print("TYPE:", dtype)
        print()
```

```python
bad = []

for file in folder.glob("*.parquet"):
    schema = pq.read_schema(file)
    
    if str(schema.field("duration_seconds").type) != "int64":
        bad.append(file)

print("Bad files:", len(bad))
```

```python
from pyspark.sql.functions import col, to_timestamp, hour, year, month, avg,col, dayofmonth
# read cleaned parquet
df = spark.read.parquet("data/clean/*.parquet")
df = df.withColumn(
    "duration_seconds",
    col("duration_seconds").cast("long")
)
df = df.withColumn(
    "start_time",
    col("start_time").cast("long")
)

df = df.filter(col("duration_seconds").isNotNull())

df.show(5)
df.printSchema()
# convert timestamp
df = df.withColumn(
    "start_time",
    to_timestamp(col("start_time"), "dd/MM/yyyy HH:mm")
)
# derive time features
df = df.withColumn("year", year("start_time"))
df = df.withColumn("month", month("start_time"))
df = df.withColumn("day", dayofmonth("start_time"))
df = df.withColumn("hour", hour("start_time"))
df.show(5)
```
